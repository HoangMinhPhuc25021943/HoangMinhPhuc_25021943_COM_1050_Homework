```python
#11
while True:
  try:
    a = int(input("Nhập cạnh a: "))
    b = int(input("Nhập cạnh b: "))
    c = int(input("Nhập cạnh c: "))

    if a > 0 and b > 0 and c > 0:
      if (a + b > c) and (a + c > b) and (b + c > a):
        if a == b == c:
           print("Tam giác đều")
        elif a == b or a == c or b == c:
          print("Tam giác cân")
        else:
          print("Tam giác thường")
    else:
      print("Không phải tam giác")
    break
  except ValueError:
    print("Bạn cần nhập vào các số nguyên dương")

```

```python
#13
while True:
   try:
    so_dien=float(input("Nhập vào số điện tiêu thụ:"))
    if so_dien >= 0:
       if so_dien <= 50:
          tien = so_dien * 1500
       elif so_dien <= 100:
          tien = 50 * 1500 + (so_dien - 50) * 2000
       else:
          tien = 50 * 1500 + 50 * 2000 + (so_dien - 100) * 3000
       break
    else:
      print("Số điện tiêu thụ phải không âm")
   except ValueError:
    print("Số điện nhập vào phải là số không âm")

print("Tiền điện phải trả là:", tien, "đồng")
```

```python
#15
while True:
    try:
        dtb = float(input("Nhập điểm trung bình: "))
        if 0 <= dtb <= 10:
            if dtb >= 8.0:
                print("Học lực: Giỏi")
            elif dtb >= 6.5:
                print("Học lực: Khá")
            elif dtb >= 5.0:
                print("Học lực: Trung bình")
            else:
                print("Học lực: Yếu")
            break
        else:
            print("điểm tb là số thực từ 0 tới 10 ")
    except ValueError:
        print(" điểm tb là số thực từ 0 tới 10")

```

```python
while True:
    try:
        so = float(input("Nhập vào một số thực: "))
        phan_nguyen = int(so)
        thap_phan = so - phan_nguyen

        if so >= 0:
            lam_tron_xuong = phan_nguyen
        else:
            if thap_phan != 0:
                lam_tron_xuong = phan_nguyen - 1
            else:
                lam_tron_xuong = phan_nguyen

        if so >= 0:
            if thap_phan != 0:
                lam_tron_len = phan_nguyen + 1
            else:
                lam_tron_len = phan_nguyen
        else:
            lam_tron_len = phan_nguyen

        if abs(thap_phan) < 0.5:
            lam_tron_gan1 = phan_nguyen
        else:
            if so > 0:
                lam_tron_gan1 = phan_nguyen + 1
            else:
                lam_tron_gan1 = phan_nguyen - 1

        print(lam_tron_len, lam_tron_xuong, lam_tron_gan1)
        break

    except ValueError:
        print("Bạn cần nhập 1 số thực")
```
