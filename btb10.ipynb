```python
#1
def merge(a,b):
    i = j = 0 
    res = []
    while i < len(a) and j < len(b): 
        if a[i] < b[j]: 
            res.append(a[i])
            i += 1
        else: 
            res.append(b[j])
            j += 1
    res = res + a[i: ]
    res = res + b[j: ]
    return res

a = [int(n) for n in input().split()]
b = [int(n) for n in input().split()]
print(merge(a,b))
```

```python
#2
def find(matrix, k): 
    m = len(matrix) # so dong
    n = len(matrix[0]) # so cot


    
    low = matrix[0][0]
    high = matrix[m-1][n-1]
    ans = low
    def countLessEqual(mid):
        count = 0
        r = m - 1  # Bắt đầu ở hàng cuối cùng
        c = 0      # Cột đầu tiên
        while r >= 0 and c < n:
            if matrix[r][c] <= mid:
                # Nếu phần tử tại (r, c) <= mid, thì tất cả các phần tử phía trên nó trong cột c cũng <= mid
                count += (r + 1)
                c += 1
            else:
                # Nếu lớn hơn, nhảy lên hàng trên
                r -= 1
        return count
    while low <= high:
        mid = (low + high) // 2
        if countLessEqual(mid) >= k:
            ans = mid
            high = mid - 1
        else:
            low = mid + 1
            
    return ans

matrix = []
k = 0
while True:
    tem = input() 
    if not tem: break # Thoát nếu dòng trống
    row = list(map(int, tem.split())) # get input to interger in 1 row
    if len(row) == 1: 
        k = row[0]
        break
    matrix.append(row)
print(find(matrix,k))
```

```python
#9 
def Median(nums):
    a, b, c, d, e = nums

    if a < b:
        min_ab, max_ab = a, b
    else:
        min_ab, max_ab = b, a

    if c < d:
        min_cd, max_cd = c, d
    else:
        min_cd, max_cd = d, c

    if min_ab < min_cd:
        small = min_cd
    else:
        small = min_ab

    if max_ab > max_cd:
        large = max_cd
    else:
        large = max_ab

    if small < e:
        if e < large:
            return e
        else:
            return large
    else:
        if small < large:
            return small
        else:
            return large

```
