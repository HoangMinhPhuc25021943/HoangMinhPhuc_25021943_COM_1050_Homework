```python
#1
def Max_Of_Two(a,b):
    if a > b:
        return a
    elif a < b:
        return b
```

```python
#2
def swap(a,b):
    a, b = b, a 
    return a, b 
print(swap(18,36)) 
```

```python
#3
def prime(n):
    s = 0
    for i in range(1, n+1):
        if n%i == 0:
            s += 1
    if s == 2:
        return "True"
    else:
        return "False"
```

```python
#4
def perfect(n):
    s = 0 
    for i in range(1,n):
        if n%i == 0:
            s+=i
    return s==n
print(perfect(6))
```

```python
#5
def find():
    a = list(map(int, input().split()))
    b =int(input())
    for i in range(len(a)):
      if a[i]==b:
       return i+1
    return -1
find()
```

```python
#6
def factorial(n):
    s=1
    for i in range(1,n+1):
        s*=i
    print(s)
factorial(6)
```

```python
#7 
def cal(a,b,c):
    if b=="+":
     return round(float(a) + float(c),2)
    elif b=="-":
     return round(float(a) - float(c),2)
    elif b=="*":
     return round(float(a) * float(c),2)
    elif b=="/":
        if c!=0:
           return round(float(a) / float(c),2)
        else:
           print("ERROR")
            
    else:
        print("SyntaxError")
a,b,c=list(map(str, input().split()))
print(cal(a,b,c))
```

```python
#10
def dang_cau(a, b):
    n = len(a)
    for i in range(n):
        for j in range(i):
            if a[i] == a[j] and b[i] != b[j]:
                return False
            if a[i] != a[j] and b[i] == b[j]:
                return False
    return True
a, b = list(map(str,input().split()))
print(dang_cau(a,b))
```
